# Amazon Sales Analysis

## 1. Environment configuration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configuración de visualización
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline
```

## 2. Load and First Exploration

```python
df = pd.read_csv('/content/amazon.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
#Se eliminan símbolos en columnas numéricas y se convierten str a float
df['discounted_price'] = df['discounted_price'].str.replace('₹|,', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹|,', '', regex=True).astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=True).astype(float)
```

```python
df.info()
```

```python
#Con un dataset de 1K+, perder 2 filas es despreciable (0.2%)
#Se eliminan las vacías
df = df.dropna(subset=['rating_count'])
print(f"Filas restantes: {len(df)}")
```

```python
#Comprobación de eliminación de filas vacías
df.info()
```

```python
# Chequeo completo de nulos
print(df.isnull().sum())
print(f"\nPorcentaje de nulos por columna:")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
#Continua exploración
df.describe()
```

## 3. Price distribution by category

```python
df['category'].unique()
```

```python
# Se extrae la primera categoría (más general) ya que es difícil observarlas sin
# jerarquía.
df['main_category'] = df['category'].str.split('|').str[0]

print("Categorías principales:")
print(df['main_category'].value_counts())

#Se observa cuántas categorías principales hay
print(f"\nNúmero de categorías principales: {df['main_category'].nunique()}")
```
