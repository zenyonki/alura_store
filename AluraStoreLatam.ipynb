# 4. Productos más y menos vendidos

```python
def productos_extremos(df):
    # Contar la cantidad de productos por categoría
    recuento = df['Categoría del Producto'].value_counts()

    # Identificar el producto más vendido
    producto_mas_vendido = recuento.idxmax()
    cantidad_mas_vendida = recuento.max()

    # Identificar el producto menos vendido
    producto_menos_vendido = recuento.idxmin()
    cantidad_menos_vendida = recuento.min()

    # Crear un diccionario con los resultados
    resultado = {
        "mas vendido": (producto_mas_vendido, cantidad_mas_vendida),
        "menos vendido": (producto_menos_vendido, cantidad_menos_vendida)
    }

    return resultado

# Aplicar la función a cada tienda
ventas_extremas_tienda_1 = productos_extremos(tienda)
ventas_extremas_tienda_2 = productos_extremos(tienda2)
ventas_extremas_tienda_3 = productos_extremos(tienda3)
ventas_extremas_tienda_4 = productos_extremos(tienda4)

# Mostrar resultados
print("Tienda 1:", ventas_extremas_tienda_1)
print("Tienda 2:", ventas_extremas_tienda_2)
print("Tienda 3:", ventas_extremas_tienda_3)
print("Tienda 4:", ventas_extremas_tienda_4)
```

```python
def proporciones_categoria(df):
    # Total de ventas de la tienda
    total_ventas = df['Categoría del Producto'].count()

    # Contar la cantidad de productos por categoría
    recuento = df['Categoría del Producto'].value_counts()

    # Identificar el producto más vendido
    producto_mas_vendido = recuento.idxmax()
    cantidad_mas_vendida = recuento.max()
    proporcion_mas_vendida = cantidad_mas_vendida / total_ventas

    # Identificar el producto menos vendido
    producto_menos_vendido = recuento.idxmin()
    cantidad_menos_vendida = recuento.min()
    proporcion_menos_vendida = cantidad_menos_vendida / total_ventas

    # Crear un diccionario con los resultados
    resultado = {
        "mas vendido": {
            "categoría": producto_mas_vendido,
            "cantidad": cantidad_mas_vendida,
            "proporcion": round(proporcion_mas_vendida, 4)  # Redondeo para claridad
        },
        "menos vendido": {
            "categoría": producto_menos_vendido,
            "cantidad": cantidad_menos_vendida,
            "proporcion": round(proporcion_menos_vendida, 4)
        }
    }

    return resultado

# Aplicar la función a cada tienda
proporciones_tienda_1 = proporciones_categoria(tienda)
proporciones_tienda_2 = proporciones_categoria(tienda2)
proporciones_tienda_3 = proporciones_categoria(tienda3)
proporciones_tienda_4 = proporciones_categoria(tienda4)

def imprimir_proporciones(tienda_resultados, tienda_nombre):
    print(f"\n{tienda_nombre}:")
    print(f"  Productos más vendido: {tienda_resultados['mas vendido']['categoría']}")
    print(f"    Cantidad: {tienda_resultados['mas vendido']['cantidad']}")
    print(f"    Proporción: {tienda_resultados['mas vendido']['proporcion'] * 100:.2f}%")
    print(f"  Productos menos vendido: {tienda_resultados['menos vendido']['categoría']}")
    print(f"    Cantidad: {tienda_resultados['menos vendido']['cantidad']}")
    print(f"    Proporción: {tienda_resultados['menos vendido']['proporcion'] * 100:.2f}%")

# Llamar la función para cada tienda
imprimir_proporciones(proporciones_tienda_1, "Tienda 1")
imprimir_proporciones(proporciones_tienda_2, "Tienda 2")
imprimir_proporciones(proporciones_tienda_3, "Tienda 3")
imprimir_proporciones(proporciones_tienda_4, "Tienda 4")
```

```python
def margen_de_beneficio(df):

    # Total de ventas por categoría (suma de 'Precio')
    ingresos_por_categoria = df.groupby('Categoría del Producto')['Precio'].sum()

    # Identificar la categoría más vendida
    categoria_mas_vendida = ingresos_por_categoria.idxmax()
    ingreso_mas_vendido = ingresos_por_categoria.max()

    # Identificar la categoría menos vendida
    categoria_menos_vendida = ingresos_por_categoria.idxmin()
    ingreso_menos_vendido = ingresos_por_categoria.min()

    # Crear un diccionario con los resultados
    resultado = {
        "mas vendido": {
            "categoría": categoria_mas_vendida,
            "ingreso": round(ingreso_mas_vendido, 2)
        },
        "menos vendido": {
            "categoría": categoria_menos_vendida,
            "ingreso": round(ingreso_menos_vendido, 2)
        }
    }

    return resultado

# Aplicar la función a cada tienda
beneficio_tienda_1 = margen_de_beneficio(tienda)
beneficio_tienda_2 = margen_de_beneficio(tienda2)
beneficio_tienda_3 = margen_de_beneficio(tienda3)
beneficio_tienda_4 = margen_de_beneficio(tienda4)

# Mostrar resultados
def imprimir_beneficio(tienda_resultados, tienda_nombre):
    print(f"\n{tienda_nombre}:")
    print(f"  Categoría más rentable: {tienda_resultados['mas vendido']['categoría']}")
    print(f"    Ingreso: ${tienda_resultados['mas vendido']['ingreso']:,.2f}")
    print(f"  Categoría menos rentable: {tienda_resultados['menos vendido']['categoría']}")
    print(f"    Ingreso: ${tienda_resultados['menos vendido']['ingreso']:,.2f}")

# Llamar la función para cada tienda
imprimir_beneficio(beneficio_tienda_1, "Tienda 1")
imprimir_beneficio(beneficio_tienda_2, "Tienda 2")
imprimir_beneficio(beneficio_tienda_3, "Tienda 3")
imprimir_beneficio(beneficio_tienda_4, "Tienda 4")
```

```python
mas_rentable = [
    beneficio_tienda_1["mas vendido"]["ingreso"],
    beneficio_tienda_2["mas vendido"]["ingreso"],
    beneficio_tienda_3["mas vendido"]["ingreso"],
    beneficio_tienda_4["mas vendido"]["ingreso"]
]
menos_rentable = [
    beneficio_tienda_1["menos vendido"]["ingreso"],
    beneficio_tienda_2["menos vendido"]["ingreso"],
    beneficio_tienda_3["menos vendido"]["ingreso"],
    beneficio_tienda_4["menos vendido"]["ingreso"]
]
total_facturacion = [
    ingreso_total_1,
    ingreso_total_2,
    ingreso_total_3,
    ingreso_total_4
]
resto = [total - mas - menos for total, mas, menos in zip(total_facturacion, mas_rentable, menos_rentable)]

# Crear el gráfico de líneas apiladas
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar áreas apiladas
ax.fill_between(tiendas, mas_rentable, label='Más Rentable', color='#09122C', alpha=1)
ax.fill_between(tiendas, [mas + menos for mas, menos in zip(mas_rentable, menos_rentable)], mas_rentable, label='Menos Rentable', color='#E17564', alpha=0.6)
ax.fill_between(tiendas, total_facturacion, [mas + menos for mas, menos in zip(mas_rentable, menos_rentable)], label='Resto', color='#BE3144', alpha=0.6)

# Etiquetas en los puntos más y menos rentables
for i, tienda in enumerate(tiendas):
    ax.text(i, mas_rentable[i] / 2, f"{beneficio_tienda_1['mas vendido']['categoría']}\n${mas_rentable[i]/1e6:,.2f}M", ha='center', va='center', color='white', fontsize=8)
    ax.text(i, (mas_rentable[i] + menos_rentable[i] / 2), f"{beneficio_tienda_1['menos vendido']['categoría']}\n${menos_rentable[i]/1e6:,.2f}M", ha='center', va='center', color='white', fontsize=8)

# Configuración de etiquetas y título
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f'${value/1e6:,.0f}M'))
ax.set_title("Facturación por Categoría en Cada Tienda", fontsize=16)
ax.set_xlabel("Tiendas", fontsize=12)
ax.set_ylabel("Ingreso Total ($)", fontsize=12)
ax.legend()

#Nota al margen
ax.annotate(
    "Nota: La 'M' representa 'millones'",
    xy=(0, 4),
    xytext=(-50, -50),
    textcoords='offset points',
    fontsize=12,
    color="white",
    ha="center",
    va="top"
)

plt.style.use('dark_background')
plt.tight_layout()
plt.show()

```
