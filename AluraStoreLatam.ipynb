#1. Análisis de facturación



```python
ingreso_total_1 = tienda['Precio'].sum()
ingreso_total_2 = tienda2['Precio'].sum()
ingreso_total_3 = tienda3['Precio'].sum()
ingreso_total_4 = tienda4['Precio'].sum()

print(f'El ingreso total de la Tienda 1 es: {ingreso_total_1:,.2f}')
print(f'El ingreso total de la Tienda 2 es: {ingreso_total_2:,.2f}')
print(f'El ingreso total de la Tienda 3 es: {ingreso_total_3:,.2f}')
print(f'El ingreso total de la Tienda 4 es: {ingreso_total_4:,.2f}')
```

```python
fig, ax = plt.subplots()

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingreso_total = [ingreso_total_1, ingreso_total_2, ingreso_total_3, ingreso_total_4]
bar_labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
#Paleta de colores personalizada de colorhunt.co
bar_colors = ['#09122C', '#872341', '#BE3144', '#E17564']

ax.bar(tiendas, ingreso_total, label=bar_labels, color=bar_colors)

ax.set_ylabel('Ingresos')
ax.set_title('Tiendas del Callenge Sr. Juan')
# Configuramos el formato del eje Y para mostrar los valores en millones (M).
# Usamos FuncFormatter de matplotlib.ticker para personalizar cómo se muestran las etiquetas del eje.
# La función lambda toma el valor 'x' del eje Y y lo divide por 1e6 (millones),
# formateando el resultado con un decimal y agregando la letra 'M' (millones).
# Ejemplo: un valor de 1,500,000 en el eje Y se mostrará como '1.5M'.
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1e6:,.1f}M')) #
ax.set_ylim(0, max(ingreso_total) * 1.2)

# Agregamos etiquetas de texto encima de cada barra para mostrar su valor.
# 'enumerate' nos da tanto el índice (i) como el valor (value) de cada barra.
# Colocamos el texto ligeramente por encima de cada barra (value + 5% del valor máximo)
# para mejorar la legibilidad, y usamos f-strings para formatear los valores con comas
# separadoras y dos decimales.
for i, value in enumerate(ingreso_total):
    ax.text(i, value + 0.05 * max(ingreso_total), f'{value/1e6:,.2f}M', ha='center', va='bottom')

plt.style.use('dark_background')

plt.show()
```

```python
#Esto es para conocer los estilos disponibles en la librería. Se probaron todos y se eligió el "dark_background".
print(plt.style.available)
```

```python
#Aquí se obtiene la proporción total de las tiendas para generar un gráfico de pastel
#se pretende que esto ayude a la toma de decisión puesto que de manera gráfica se
#podrá observar qué tienda aporta más al Sr. Juan

ingreso_total_general = sum(ingreso_total)
proporciones = [ingreso / ingreso_total_general for ingreso in ingreso_total]
proporciones_porcentaje = [f'{proporcion * 100:.1f}%' for proporcion in proporciones]
for i, (ingreso, porcentaje) in enumerate(zip(ingreso_total, proporciones_porcentaje), start=1):
    print(f'Tienda {i}: {porcentaje} del ingreso total')
```

```python
colors = ['#09122C', '#872341', '#BE3144', '#E17564']

fig, ax = plt.subplots()
ax.pie(
    proporciones,
    labels=tiendas,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'black'}
)

ax.set_title('Proporción de ingresos por tienda')

plt.style.use('dark_background')

plt.show()
```
